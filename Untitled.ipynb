```python


import pandas as pd
import numpy as np
import pyodbc
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

output_dir = "rfid_etl_output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

```

```python

print("1. EXTRACT PHASE - Starting data extraction...")
CONNECTION_STRING = (
    'DRIVER={ODBC Driver 18 for SQL Server};'
    'SERVER=IHEB;'
    'DATABASE=rfid;'
    'Trusted_Connection=Yes;'
    'Encrypt=no;'
)

try:
    conn = pyodbc.connect(CONNECTION_STRING)
    print("Database connection successful")
except Exception as e:
    print(f"Error connecting to database: {e}")
    exit(1)

tables = {
    "Date": "SELECT * FROM Date",
    "Equipe": "SELECT * FROM Equipe",
    "Poste_Competence": "SELECT * FROM Poste_Competence",
    "Employe": "SELECT * FROM Employe",
    "Alerte": "SELECT * FROM Alerte",
    "Evenement": "SELECT * FROM Evenement"
}

dfs = {}
for table_name, query in tables.items():
    try:
        dfs[table_name] = pd.read_sql(query, conn)
        print(f"Extracted {len(dfs[table_name])} rows from {table_name}")
    except Exception as e:
        print(f"Error extracting {table_name}: {e}")

conn.close()

print("Saving raw extracted data...")
for table_name, df in dfs.items():
    df.to_csv(f"{output_dir}/raw_{table_name}.csv", index=False)

```

```python
print("\n2. TRANSFORM PHASE - Starting data transformation...")

date_columns = {
    "Date": ["date_complete"],
    "Employe": ["date_naissance", "date_embauche"],
    "Alerte": ["date_alerte"],
    "Evenement": ["date_evenement"]
}

for table, cols in date_columns.items():
    if table in dfs:
        for col in cols:
            if col in dfs[table].columns:
                dfs[table][col] = pd.to_datetime(dfs[table][col])

print("Creating enhanced employee dataset...")
if all(k in dfs for k in ["Employe", "Equipe", "Poste_Competence"]):
    employees_enhanced = dfs["Employe"].merge(
        dfs["Equipe"], 
        left_on='equipe_id', 
        right_on='equipe_id',
        how='left'
    )
    
    employees_enhanced = employees_enhanced.merge(
        dfs["Poste_Competence"], 
        left_on='poste_id', 
        right_on='poste_id',
        how='left'
    )
    
    employees_enhanced['tenure_years'] = ((datetime.now() - employees_enhanced['date_embauche'])
                                           .dt.days / 365).round(1)
    
    dfs["Employe_Enhanced"] = employees_enhanced
    print(f"Created enhanced employee dataset with {len(employees_enhanced)} rows")

print("Creating enhanced alerts dataset...")
if all(k in dfs for k in ["Alerte", "Employe_Enhanced"]):
    alerts_enhanced = dfs["Alerte"].merge(
        dfs["Employe_Enhanced"][['rfid', 'nom', 'prenom', 'nom_equipe', 'titre_poste']],
        on='rfid',
        how='left'
    )
    
    alerts_enhanced['alert_month'] = alerts_enhanced['date_alerte'].dt.month
    alerts_enhanced['alert_year'] = alerts_enhanced['date_alerte'].dt.year
    alerts_enhanced['alert_day_of_week'] = alerts_enhanced['date_alerte'].dt.day_name()
    
    dfs["Alerte_Enhanced"] = alerts_enhanced
    print(f"Created enhanced alerts dataset with {len(alerts_enhanced)} rows")

print("Creating enhanced events dataset...")
if all(k in dfs for k in ["Evenement", "Employe_Enhanced"]):
    events_enhanced = dfs["Evenement"].merge(
        dfs["Employe_Enhanced"][['rfid', 'nom', 'prenom', 'nom_equipe', 'titre_poste']],
        on='rfid',
        how='left'
    )

    events_enhanced['event_month'] = events_enhanced['date_evenement'].dt.month
    events_enhanced['event_year'] = events_enhanced['date_evenement'].dt.year
    events_enhanced['event_day_of_week'] = events_enhanced['date_evenement'].dt.day_name()
    events_enhanced['event_hour'] = events_enhanced['date_evenement'].dt.hour
    
    dfs["Evenement_Enhanced"] = events_enhanced
    print(f"Created enhanced events dataset with {len(events_enhanced)} rows")

print("Creating team statistics...")
if "Employe_Enhanced" in dfs:
    team_stats = dfs["Employe_Enhanced"].groupby('nom_equipe').agg({
        'rfid': 'count',
        'tenure_years': 'mean'
    }).rename(columns={
        'rfid': 'employee_count',
        'tenure_years': 'avg_tenure'
    }).reset_index()
    
    if "Alerte_Enhanced" in dfs:
        team_alerts = dfs["Alerte_Enhanced"].groupby('nom_equipe').size().reset_index(name='alert_count')
        team_stats = team_stats.merge(team_alerts, on='nom_equipe', how='left')
        team_stats['alerts_per_employee'] = team_stats['alert_count'] / team_stats['employee_count']
    
    dfs["Team_Stats"] = team_stats
    print(f"Created team statistics with {len(team_stats)} rows")

print("Creating attendance analysis...")
if "Evenement_Enhanced" in dfs:
    entry_exit = dfs["Evenement_Enhanced"][
        dfs["Evenement_Enhanced"]['type_evenement'].isin(['Entr√©e', 'Sortie'])
    ]
    
    daily_attendance = entry_exit.groupby(
        [entry_exit['date_evenement'].dt.date, 'rfid', 'type_evenement']
    ).agg({
        'date_evenement': ['min', 'count']
    }).reset_index()
    
    daily_attendance.columns = ['date', 'rfid', 'event_type', 'first_time', 'event_count']
    
    attendance_pivot = daily_attendance.pivot_table(
        index=['date', 'rfid'],
        columns='event_type',
        values=['first_time', 'event_count']
    ).reset_index()

    attendance_pivot.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in attendance_pivot.columns]
    
    dfs["Daily_Attendance"] = attendance_pivot
    print(f"Created attendance analysis with {len(attendance_pivot)} rows")

```

```python
print("\n3. LOAD PHASE - Saving transformed data...")

for name, df in dfs.items():
    if name not in ["Date", "Equipe", "Poste_Competence", "Employe", "Alerte", "Evenement"]:
        output_path = f"{output_dir}/{name}.csv"
        df.to_csv(output_path, index=False)
        print(f"Saved {name} to {output_path}")

print("\n4. Creating basic visualizations...")

if "Team_Stats" in dfs:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='employee_count', y='nom_equipe', data=dfs["Team_Stats"])
    plt.title('Employee Count by Team')
    plt.tight_layout()
    plt.savefig(f"{output_dir}/team_employee_count.png")
    print(f"Saved team employee count visualization")

if "Alerte_Enhanced" in dfs:
    plt.figure(figsize=(10, 6))
    alert_counts = dfs["Alerte_Enhanced"]['type_alerte'].value_counts()
    sns.barplot(x=alert_counts.index, y=alert_counts.values)
    plt.title('Alert Types Distribution')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(f"{output_dir}/alert_types.png")
    print(f"Saved alert types visualization")

if "Evenement_Enhanced" in dfs:
    plt.figure(figsize=(10, 6))
    events_by_hour = dfs["Evenement_Enhanced"]['event_hour'].value_counts().sort_index()
    sns.barplot(x=events_by_hour.index, y=events_by_hour.values)
    plt.title('Events by Hour of Day')
    plt.xticks(range(0, 24, 2))
    plt.tight_layout()
    plt.savefig(f"{output_dir}/events_by_hour.png")
    print(f"Saved events by hour visualization")

print("\nETL process completed successfully!")
print(f"All output files saved to {output_dir}/")
```
